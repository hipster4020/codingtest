# *피드 포워드 신경망(Feed-Forward Neural Network, FFNN)*


<br><b>순방향 신경망</b>(Feed-Forward Neural Network)<br>
다층 퍼셉트론(MLP)과 같이 오직 입력층에서 출력층 방향으로 연산이 전개되는 신경망을 피드포워드 신경망(Feed-Forward NeuralNetwork, FFNN)이라고 함


<br><b>순환 신경망</b>(Recurrent Neural Network)<br>
FFNN에 속하지 않는 RNN이라는 신경망<br>
이 신경망은 <b>은닉층의 출력값을 출력층</b>으로도 값을 보내지만, 동시에 <b>은닉층의 출력값이 다시 은닉층의 입력</b>으로 사용

# 

# *전결합층(Fully-connected Layer, FC, Dense Layer)*

다층 퍼셉트론은 은닉층과 출력층에 있는 모든 뉴런은 바로 이전 층의 모든 뉴런과 연결되어 있다<br>
<u>어떤 층의 모든 뉴런이 이전 층의 모든 뉴런과 연결되어 있는 층</u>을 <big><b>전결합층(Fully-connected Layer)</b></big>라고 함
