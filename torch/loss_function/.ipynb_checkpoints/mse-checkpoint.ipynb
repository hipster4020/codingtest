### *신경망의 출력(y hat)과 타깃(y)이 연속값인 회귀 문제에서 사용됨*
### *평균 제곱 오차(mse)는 예측과 타깃값 차이를 제곱하여 평균한 값*
### *종류 : MSE(Mean Squared Error), MAE(Mean Absolute Error), RMSE(Root Mean Squared Error)*
<br>



```python
import torch
import torch.nn as nn
```

```python
mse_loss = nn.MSELoss()
```

```python
outputs = torch.randn(3, 5, requires_grad = True)
targets = torch.randn(3, 5)
loss = mse_loss(outputs, targets)
print(loss)
```
