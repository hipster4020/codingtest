# *경사 하강법*

경사 하강법은 비용 함수를 미분하여 기울기(Gradient)를 구해서 오차를 최소화하는 방향을 찾아내는 알고리즘<br>


# 

# *자동  미분(Autograd) 실습*

```python
import torch
```

```python
w = torch.tensor(2.0, requires_grad=True)
```

값이 2인 스칼라 텐서 w를 선언<br>
required_grad=True : 텐서에 대한 기울기 설정<br>

```python
y = w ** 2
z = 2*y + 5
```

```python
z.backward()
```

```python
print(f'수식을 w로 미분한 값 : {w.grad}')
```
