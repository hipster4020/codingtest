```python
import torch
```

```python
x_train = torch.Tensor([[1], [2], [3]])
y_train = torch.Tensor([[2], [4], [6]])
```

# 

### *가설(Hypothesis) 수립*

머신러닝에서 세우는 식을 <b>가설</b>이라고 함<br>
가설은 임의로 추측해서 세우는 식일 수도 있고, 경험적으로 알고 있는 식일 수도 있음<br>
맞는 가설이 아니라고 판단되면 계속 수정해나가게 됨

<br><big><big><b>선형 회귀</b></big></big><br>
학습 데이터와 가장 잘 맞는 하나의 직선을 찾는 일<br>


x와 곱해지는 <b>W를 가중치(Weight)</b>라고 하며, <b>b를 편향(bias)</b>라고 함<br>

# 

# *비용 함수(Cost Function)*

<b>비용 함수(Cost Function) = 손실 함수(Loss Function) = 오차 함수(Error Function) = 목적 함수(Objective Function)</b>


4개의 점을 가장 잘 표현하는 직선을 찾을 때, <b>직선(예측값)과 점들(실제값)과의 총 오차를 통해 근거를 찾음</b>

### 

### 실제값과 예측값(주황색 선)과의 오차 비교표



### MSE(Mean Squared Error)
&nbsp;&nbsp;&nbsp;각각의 모든 오차에 대해 제곱 후 더하고 갯수로 나눈다


# 

# *👉🏻 옵티마이저*

<b>옵티마이저(최적화 알고리즘)</b>를 통해 비용 함수의 값을 최소로 하는 W와 b를 찾는다<br>
옵티마이저 알고리즘을 통해 적절한 W와 b를 찾아내는 과정을 머신러닝에서 <b>학습</b>이라고 함

### *⭐️ 경사 하강법(Gradient Descent)*


주황색 선은 기울기 W가 20으로 y = 20x, 녹색 선은 기울기가 W가 1로 y = x<br>
↕는 <b>각 점에서의 실제값과 두 직선의 예측값의 오차</b><br>
즉, 기울기가 지나치게 크면 실제값과 예측값의 오차가 커지고, 기울기가 지나치게 작아도 실제값과 예측값의 오차가 커짐


기울기 W가 무한대로 커지면 <b>커질수록 cost 값 또한 무한대로 커지고</b>, 반대로 기울기 W가 무한대로 <b>작아져도 cost 값이 무한대로 커짐</b><br>
### 기계가 해야할 일은 cost가 가장 최소값을 가지게 하는 W를 찾는 것으로, 맨 아래의 볼록한 부분의 기울기 W를 찾아야 함


<br>기계는 <big><b>경사 하강법</b></big>을 통해 임의의 초기값 W값을 정한 뒤, 맨 아래의 볼록한 부분을 향해 점자 W값을 수정<br><br>

## *경사 하강법*

#### cost가 최소가 되는 지점은 접선 기울기가 0이 되는 지점이며, 미분값이 0이 되는 지점
#### 비용 함수를 미분하여 현재 W에서의 접선의 기울기를 구하고, 접선의 기울기가 낮은 방향으로 W의 값을 변경하는 작업을 반복



반복 작업에서 현재 W에 접선의 기울기를 구해 특정 숫자 α를 곱한 값을 빼서 새로운 W로 사용하는 식이 사용<br>
기울기가 음수던지 양수던지 결과적으로 <b>기울기가 0인 방향으로 W값이 조정됨</b><br><br>

### 학습률(Learning Rate) α
학습률 α는 W값을 변경할 때, 얼마나 크게 변경할지 결정<br>
또한, W를 그래프의 한 점으로 보고 접선의 기울기가 0일 때까지 경사를 따라 내려가는 관점으로 얼마나 큰 폭으로 이동할지 결정<br>
* 직관적으로 학습률 α를 크게하여 기울기가 최소값이 되는 W를 빠르게 찾는 것은 좋지 않다


### 실제 경사 하강법은 W와 b에 대해 동시에 수행하여 최적의 W와 b를 찾아냄

# 

# *👉🏻 선형 회귀 구현*

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# 동일한 가중치가 나오도록 random seed 설정
torch.manual_seed(1)
```

```python
x_train = torch.Tensor([[1], [2], [3]])
y_train = torch.Tensor([[2], [4], [6]])
```

### *가중치, 편향 초기화*

```python
W = torch.zeros(1, requires_grad=True)
print(f"W : {W}")
b = torch.zeros(1, requires_grad=True)
print(f"b : {b}")
```

<big><big><b>⭐️ requires_grad</b></big></big>는 <b>자동 미분 설정</b><br><br>
required_grad=True가 적용된 텐서에 연산을 하면, 계산 그래프가 생성되며<br>
<b>backward</b> 함수를 호출하면 그래프로부터 자동으로 미분이 계산됨

### 

### *가설 세우기*

```python
hypothesis = x_train * W + b
print(hypothesis)
```

### 

### *비용 함수*
평균 제곱 오차<br>


```python
cost = torch.mean((hypothesis - y_train) ** 2)
print(cost)
```

### 

### *경사 하강법*

```python
optimizer = optim.SGD([W, b], lr = 0.01)
print(optimizer)
```

```python
# gradient를 0으로 초기화
optimizer.zero_grad() 
# 비용 함수를 미분하여 gradient 계산
cost.backward() 
# W와 b를 업데이트
optimizer.step()
```

### 

### *⭐️ 전체 코드*

```python
# 데이터
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
# 모델 초기화
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
# optimizer 설정
optimizer = optim.SGD([W, b], lr=0.01)

nb_epochs = 1999 # 원하는만큼 경사 하강법을 반복
for epoch in range(nb_epochs + 1):

    # H(x) 계산
    hypothesis = x_train * W + b

    # cost 계산
    cost = torch.mean((hypothesis - y_train) ** 2)

    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    # 100번마다 로그 출력
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            epoch, nb_epochs, W.item(), b.item(), cost.item()
        ))
```

최적의 기울기 W가 2에 가깝고, b는 0에 가까워졌다<br>
실제 정답은 H(x) = 2x이므로 거의 정답이라고 할 수 있다.
