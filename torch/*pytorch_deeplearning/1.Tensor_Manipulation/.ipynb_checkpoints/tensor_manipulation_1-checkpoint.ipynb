# *Numpy & Tensor*

```python
import numpy as np
import torch
```

### *1D with Numpy*

```python
t1 = np.array([0, 1, 2, 3, 4, 5, 6])
t2 = np.array([0., 1., 2., 3., 4., 5., 6.])
print(t1)
print(t2)
print('Rank of t: ', t2.ndim)
print('Shape of t: ', t2.shape)
```

### 1D with Pytorch

```python
t1 = torch.FloatTensor([0., 1., 2., 3., 4., 5., 6.])
t2 = torch.Tensor([0., 1., 2., 3., 4., 5., 6.])
print(t1.type())
print(t2.type())
print(f"dimension : {t2.dim()}")
print(f"shape : {t2.shape}")
print(f"size : {t2.size()}")
```

### *2D with Numpy*

```python
t3 = np.array([[1., 2.], [3., 4.], [5., 6.]])
print(t3)
print('Rank of t: ', t3.ndim)
print('Shape of t: ', t3.shape)
```

```python
t3 = torch.Tensor([[1., 2.], [3., 4.], [5., 6.]])
print(t3)
print(f"dimension : {t3.dim()}")
print(f"shape : {t3.shape}")
print(f"size : {t3.size()}")
```

##### *slicing*

```python
print(t3[0])
print(t3[:, 0])
print(t3[:, 1])
```

# *Broadcasting*

   ,      , PyTorch      .

### *Plus*

```python
#   
m1 = torch.Tensor([1, 2, 3, 4, 5])
m2 = torch.Tensor([2, 2, 2, 2, 2])
m = m1 + m2
print(m)
```

```python
#   
m1 = torch.Tensor([1, 2, 3])
m2 = torch.Tensor([2]) # [2]  [2, 2, 2]
m = m1 + m2
print(m)
```

```python
m1 = torch.Tensor([[1, 2]]) # 1 * 2      [[1, 2]]  [[1, 2], [1, 2]]
m2 = torch.Tensor([[3], [4]]) # 2 * 1    [[3], [4]]  [[3, 3], [4, 4]]
print(m1)
print(m1.dim())
print(m2)
print(m2.dim())

m = m1 + m2
print(m)
```

### *Matrix Multiplication*

```python
m1 = torch.Tensor([[1, 2], [3, 4]]) # 2 * 2
m2 = torch.Tensor([[1], [2]])         # 2 * 2
m = m1.matmul(m2)
print(m)
```

### *Multiplication*

```python
mat1 = m1 * m2
mat2 = m1.mul(m2)
print(mat1)
print(mat2)
```

### *Mean*

```python
t1 = torch.Tensor([1, 2])
print(t1.mean())
t2 = torch.Tensor([[1, 2, 3], [2, 3, 4]])
print(t2.mean())
#   
print(t2.mean(dim=0))
print(t2.mean(dim=1))
print(t2.mean(dim=-1))
```

### *Sum*

```python
t = torch.Tensor([[1, 2], [3, 4]])
print(t)
print(t.sum())
print(t.sum(dim=0))
print(t.sum(dim=1))
print(t.sum(dim=-1))
```

### *Max, ArgMax*

```python
t = torch.Tensor([[1, 2], [3, 4], [9, 2]])
print(t)
print(t.max())
print(t.max(dim=0))  # indices : index 
print(t.max(dim=1))
print(t.max(dim=-1))
print(" ")
print('Max: ', t.max(dim=0)[0])
print('Argmax: ', t.max(dim=0)[1])
print(" ")
print("ArgMax")
for i in t:
    print(i.argmax())
```
