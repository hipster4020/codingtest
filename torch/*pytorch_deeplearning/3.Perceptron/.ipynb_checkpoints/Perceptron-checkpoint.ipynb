# *퍼셉트론*

프랑크 로젠블라트(Frank Rosenblatt)가 1957년에 제안한 초기 형태의 인공 신경망으로 다수 입력으로부터 하나의 결과를 내보내는 알고리즘<br>
퍼셉트론은 실제 뇌를 구성하는 신경 세포 뉴련의 동작과 유사하다.<br>
뉴런은 가지돌기에서 신호를 받아들이고, 이 신호가 일정치 이상의 크기를 가지면 축삭돌기를 통해서 신호를 전달



<center>x : 신경 세포 뉴런 입력 신호 : 입력 값</center>
<center>W : 가중치, 신호를 전달하는 축삭돌기 역할</center>
<center>y : 출력 값</center>

각 입력값이 가중치와 곱해져 인공 뉴런에 보내지는데, 가중치의 값이 크면 클수록 해당 입력 값이 중요하다는 의미<br>
입력값과 가중치의 곱의 전체 합이 임계치(threshold)를 넘으면 종착지에 있는 인공 뉴런은 출력 신호로서 1을 출력하고, 그렇지 않을 경우에는 0을 출력<br>
이 함수를 계단 함수(Step function)이라고 함


세타(Θ) : 계단 함수에 사용된 임계치 값



위 식에서 임계치를 좌변으로 넘기고 편향 b(bias)로 표현하여 퍼셉트론의 입력으로 사용<br>
입력값이 1로 고정
# 

### 뉴런에서 출력값을 변경시키는 함수를 활성화 함수(Activation Function)이라고 함
### ex) 계단 함수, 시그모이드 함수, 소프트맥스 함수 등

# *단층 퍼셉트론(Single-Layer Perceptron)*


<br>
<br>
단층 퍼셉트론은 값을 보내는 단계와 값을 받아서 출력하는 두 단계로만 구성<br>
AND, NAND, OR 게이트 구현 가능<br>

### *AND Gate*


두 개의 입력 값이 모두 1인 경우에만 출력이 1이 나오는 구조

```python
def AND_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.7
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
```

```python
print(AND_gate(0, 0))
print(AND_gate(1, 0))
print(AND_gate(0, 1))
print(AND_gate(1, 1))
```

<b>NAND</b> : AND 반대<br>
    0, 0 → 1,<br>
    0, 1 → 1,<br>
    1, 0 → 1,<br>
    1, 1 → 0<br><br>
<b>OR</b> : 두 개의 입력이 모두 0인 경우에만 0<br>
    0, 0 → 0,<br>
    0, 1 → 1,<br>
    1, 0 → 1,<br>
    1, 1 → 1<br><br>
<b>XOR</b> : 두 개의 입력 모두 서로 다른 값을 가질 때만 출력 값 1, 같으면 0<br>
    0, 0 → 0,<br>
    0, 1 → 1,<br>
    1, 0 → 1,<br>
    1, 1 → 0<br>

# *다층 퍼셉트론(MultiLayer Perceptron, MLP)*


AND, NAND, OR 게이트를 조합하여 XOR 게이트를 구현한 다층 퍼셉트론<br><br>
다층 퍼셉트론 : 은닉층이 1개 이상인 퍼셉트론<br>
입력층과 출력층만 존재하는 단층 퍼셉트론과 달리 중간에 층(은닉층, Hidden Layer)을 더 추가함<br>


### 


<br>은닉층이 2개 이상인 신경망을 <b>심층 신경망(Deep Neural Network, DNN)</b>이라고 함
