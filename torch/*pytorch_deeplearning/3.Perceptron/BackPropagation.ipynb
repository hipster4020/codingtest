# *인공 신경망(Neural Network)*


<br>
입력층, 은닉층, 출력층으로 구성된 인공신경망<br>
입력층 2개, 은닉층 뉴런 2개, 출력층 2개<br>
은닉층과 출력층의 모든 뉴런은 활성화 함수로 시그모이드 사용

<b>은닉층과 출력층의 변수 z</b>는 이전층의 모든 입력이 각각의 가중치와 곱해진 값들이 모두 더해진 <b>가중합</b>을 의미<br>
<b>뉴런에서 아직 시그모이드를 거치지 않은 상태(활성화 함수의 입력)</b><br><br>
z 우측의 |를 지나서 존재하는 변수 h 또는 o는 z가 시그모이드 함수를 지난 후의 값으로 <b>각 뉴런의 출력값</b>을 의미

# 

# *순전파(Forward Propagation)*

파란색 : 입력값, 빨간색: 각 가중치

각 입력은 입력층에서 은닉층 방향으로 향하면서 각 입력에 해당하는 가중치와 곱해지고 결과적으로 가중합으로 계산되어 <b>은닉층 뉴런의 시그모이드 함수의 입력값</b>(z1, z2)<br>

z1, z2가 각각의 은닉층 뉴런에서 시그모이드 함수를 지나 리턴받은 결과값은 <b>은닉층 뉴런의 출력값</b>(h1, h2)

h1, h2는 각각 출력층의 뉴런에서 가중치와 곱해져 가중합이 되어 <b>출력층 뉴런의 시그모이드 입력값</b>이 됨(z3, z4)<br>

z3, z4가 출력층 뉴런에서 시그모이드 함수를 지난 값은 <b>해당 인공신경망이 최종적으로 계산한 출력값(예측값)</b><br>


### *이제 위 예측값과 실제값을 토대로 오차 함수를 결정함*
오차(Error)를 계산하기 위한 손실 함수(Loss Function)로 평균 제곱 오차(MSE)를 사용<br>
target: 실제값, output: 예측값<br>

# *역전파(BackPropagation)*

순전파가 입력층에서 출력층으로 향했다면 역전파는 반대로 <b>출력층에서 입력층으로 계산하면서 가중치를 업데이트하는 것</b>

##### 1단계 : 출력층 바로 이전의 은닉층을 N층이라고 할 때, 출력층과 N층 사이의 가중치를 업데이트

W5, W6, W7, W8 가중치 업데이트 진행<br><br><br>
<b>경사 하강법을 수행하기 위한 W5 가중치에 대한 업데이트 식</b><br>

1번째 항

2번째 항

3번째 항

우변의 모든 항 곱 계산

경사 하강법을 통한 가중치 업데이트

같은 원리로 W6, W7, W8 계산


##### 2단계 : N층과 N층 이전층 사이의 가중치를 업데이트
1단계와 같은 원리로 W1, W2, W3, W4에 가중치 업데이트<br>
은닉층이 여러개이면 step이 은닉층 만큼 증가<br>




업데이트 된 가중치에 대해 다시 한 번 순전파를 진행하여 오차가 감소하였는지 확인<br>

기존 오차는 0.02397190이므로 1번의 역전파로 오차가 감소함<br>
