```python
n = 5
move = ['R', 'R', 'R', 'U', 'D', 'D']
```

```python
def main(n, move):
    pos = [1, 1]
    root = [[-1, 0], [1, 0], [0, -1], [0, 1]]
    move_type = ['U', 'D', 'L', 'R']

    for i in move:
        for k, v in enumerate(move_type):
            if i == v:
                pos[0] += root[k][0]
                pos[1] += root[k][1]
                print(f"pos : {pos}")
            if pos[0] < 1 or pos[0] > n or pos[1] < 1 or pos[1] > n:
                pos[0] -= root[k][0]
                pos[1] -= root[k][1]
                continue
    return pos[0], pos[1]
```

```python
result1, result2 = main(n, move)
```

```python
print(f"result : {result1}, {result2}")
```
