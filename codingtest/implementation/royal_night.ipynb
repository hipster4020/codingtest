# 왕실의 나이트




```python
n = 'a1'
```

```python
def main(n):
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
    move = [[-1, 2], [1, 2], [2, -1], [2, 1]]
    result = 0
    
    for k, v in enumerate(alphabet):
        if v == n[0]:
            for m in move:
                dx = (k + 1) + m[0]
                dy = int(n[1]) + m[1]
                
                if dx <= 0 or dx > 8 or dy <= 0 or dy > 8:
                    continue
                result += 1
    return result
```

```python
result = main(n)
```

```python
print(f"result : {result}")
```
