```python
n, m = 4, 4
pos = [1, 1, 0]
graph = [
    [1, 1, 1, 1],
    [1, 0, 0, 1],
    [1, 1, 0, 1],
    [1, 1, 1, 1]
]
```

```python
def main(n, m, pos, graph):
    # , , , 
    direction = [0, 1, 2, 3]
    root = [[-1, 0], [0, 1], [1, 0], [0, -1]]
    check = [[False for _ in range(m)] for _ in range(n)]
    result = 0
    
    for d in direction:
        pos[2] = direction[pos[2] - 1]
        pos[0] += root[pos[2]][0]
        pos[1] += root[pos[2]][1]
        print(f"pos1 : {pos}")
        
        if pos[0] <= 0 or pos[0] > n or pos[1] <= 0 or pos[1] > m:
            print(f"out pos : {pos}")
            pos[0] -= root[pos[2]][0]
            pos[1] -= root[pos[2]][1]
            continue
        if check[pos[0]][pos[1]] == False and graph[pos[0]][pos[1]] == 0:
            check[pos[0]][pos[1]] = True
            result += 1
            print(f"if pos : {pos}") 
            pos[2] = direction[pos[2] + 1]
        else:
            print(f"else pos : {pos}")
            pos[0] -= root[pos[2]][0]
            pos[1] -= root[pos[2]][1]


    return result
```

```python
result = main(n, m, pos, graph)
```
