```python
n = 5
```

```python
def main(n):
    result = 0

    # hour
    for h in range(n+1):
        # minute
        for m in range(60):
            # second
            for s in range(60):
                if '3' in str(h)+str(m)+str(s):
                    result += 1
    return result
```

```python
result = main(n)
```

```python
print(f"result : {result}")
```
