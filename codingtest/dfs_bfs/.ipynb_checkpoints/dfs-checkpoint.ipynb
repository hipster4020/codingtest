#  *DFS    *

1.      <u>        </u><br>
2.               <br>
        .<br>
3. 2        

```python
#    DFS
def dfs(graph, v, visited):
    #    
    visited[v] = True
    print(v, end = ' ')
    
    #       
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
```

```python
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9

print(f"graph : {graph}")
print(f"visited : {visited}")
```

```python
dfs(graph, 1, visited)
```

# *BFS*
<b>Breadth First Search   </b><br>
   <br>
<u>      </u>

#  *DFS    *

1.       .
2.                
3. 2        .

BFS <b> </b>    <br>
 DFS  

```python
from collections import deque

def bfs(graph, start, visited):
    # Queue  deque 
    queue = deque([start])
    
    #    
    visited[start] = True
    
    # queue   
    while queue:
        print(queue)
        v = queue.popleft()
        print(v, end = ' ')
        
        #   ,     
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
```

```python
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

visited = [False] * 9

print(f"graph : {graph}")
print(f"visited : {visited}")
```

```python
bfs(graph, 1, visited)
```
