```python
n = int(input())


for i in range(n):
    stack = []
    data = input()
    
    for d in data:
        if d == '(':
            stack.append(d)
        elif d == ')':
            if stack:
                stack.pop()
            else:
                print("NO")
                break
    else:
        if not stack:
            print("YES")
        else:
            print("NO")
```
