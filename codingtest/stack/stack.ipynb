# 스택



```python
import sys
n = int(sys.stdin.readline())

stack=[]
for i in range(n):
    type = sys.stdin.readline()
    if 'push' in type:
        stack.append(type.split()[1])
    elif 'pop' == type:
        if not stack:
            print(-1)
        else:
            print(stack.pop())
    elif 'size' == type:
        print(len(stack))
    elif 'empty' == type:
        if not stack:
            print(1)
        else:
            print(0)
    elif 'top' == type:
        if not stack:
            print(-1)
        else:
            print(stack[-1])
```
