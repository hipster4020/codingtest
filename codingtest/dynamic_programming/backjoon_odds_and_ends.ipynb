# 설탕 배달


```python
n = int(input())

sum = 0 
while n >= 0:
    if n % 5 == 0:
        sum += int(n // 5)
        print(sum)
        break
    n -= 3
    sum += 1
else:
    print(-1)
```

```python
n = int(input())

dp = [0] * (n+1)

for i in range(2, n+1):
    dp[i] = dp[i-1] + 1
    
    if i % 2 == 0:
        dp[i] = min(dp[i], dp[i//2] + 1)
    if i % 3 == 0:
        dp[i] = min(dp[i], dp[i//3] + 1)
print(dp[n])
```

# 부녀회장이 될테야



```python
# t = int(input())

# for _ in range(t):
#     floor = int(input()) # 층
#     num = int(input())   # 호
#     f0 = [x for x in range(1, num + 1)]

#     for k in range(floor):
#         for i in range(1, num):
#             f0[i] += f0[i-1]
#     print(f0[-1])

t = int(input())

for _ in range(t):  
    floor = int(input())  # 층
    num = int(input())  # 호
    f0 = [x for x in range(1, num+1)]  # 0층 리스트
    for k in range(floor):  # 층 수 만큼 반복
        for i in range(1, num):  # 1 ~ n-1까지 (인덱스로 사용)
            f0[i] += f0[i-1]  # 층별 각 호실의 사람 수를 변경
    print(f0[-1])  # 가장 마지막 수 출력
```

# 계단 오르기



```python
N = int(input())

li = [0]*301
dp = [0]*301

for i in range(N):
    li[i] = int(input())

dp[0] = li[0]
dp[1] = li[0]+li[1]
dp[2] = max(li[0]+li[2], li[1]+li[2])
for i in range(3, N):
    dp[i] = max(dp[i-3]+li[i-1]+li[i], dp[i-2]+li[i])
print(dp[N-1])
```

# 2×n 타일링


#### 피보나치 수열 형태
#### f(n) = f(n-1) + f(n-2)


```python
n = int(input())

cache = [0] * 1001
cache[1] = 1
cache[2] = 2

for i in range(3, 1001):
    cache[i] = (cache[i-1] + cache[i-2]) % 10007
print(cache[n])
```
