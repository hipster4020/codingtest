# 미로 탐색



```python
from collections import deque

nm = input()
n = int(nm.split(' ')[0])
m = int(nm.split(' ')[1])

graph = [list(map(int, list(input()))) for _ in range(int(n))]

root = [[-1, 0], [1, 0], [0, -1], [0, 1]]

x, y = 0, 0
queue = deque()
queue.append((x, y))

while queue:
    x, y = queue.popleft()

    for i in range(len(root)):
        dx = x + root[i][0]
        dy = y + root[i][1]
        
        if dx < 0 or dx >= n or dy < 0 or dy >= m:
            continue
        if graph[dx][dy] == 0:
            continue
        if graph[dx][dy] == 1:
            graph[dx][dy] = graph[x][y] + 1
            queue.append((dx, dy))
print(graph[n - 1][m - 1])
```
