# 카드 정렬하기



```python
import heapq

cards = []
result = 0

n = int(input())
for _ in range(n):
    data = int(input())
    heapq.heappush(cards, data)

if len(cards) == 1:
    print(0)
else:
    while len(cards) > 1:
        sum = heapq.heappop(cards) + heapq.heappop(cards)
        result += sum
        heapq.heappush(cards, sum)
    print(result)
```
