```python
s = 'abac'
```

```python
list(s)
```

```python
from itertools import product
```

```python
check = list()
```

```python
[check.append(''.join(i)) for i in list(product(list(s), repeat=1))]
[check.append(''.join(i)) for i in list(product(list(s), repeat=2))]
[check.append(''.join(i)) for i in list(product(list(s), repeat=3))]
[check.append(''.join(i)) for i in list(product(list(s), repeat=4))]
```

```python
check
```

```python
check2 = [''.join(set(list(i))) for i in check]
```

```python
check2
```

```python
set(check2)
```
