# *숫자 카드 게임*




```python
# n = 3 # 행의 개수
# m = 3 # 열의 개수
# data = [
#     [3, 1, 2],
#     [4, 1, 4],
#     [2, 2, 2]
# ]

n = 2
m = 4
data = [
    [7, 3, 1, 8],
    [3, 3, 3, 4]
]
```

```python
def main(n, m, data):
    result = 0
    
    for j in data:
        min_value = min(j)
        result = max(result, min_value)

    return result
```

```python
result = main(n, m, data)
```

```python
print(result)
```
