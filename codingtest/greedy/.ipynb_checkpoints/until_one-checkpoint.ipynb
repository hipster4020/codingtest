# *1이 될때까지*

##### 데이터 입력

```python
n, k = map(int, input().split())
```

```python
result = 0
```

```python
while True:
    target = (n // k) * k
    result += (n - target)
    n = target
    print(f"target : {target}")
    print(f"result : {result}")
    
    if n < k:
        break
    
    # k로 나누기
    result += 1
    n //= k
```

##### 마지막 남은 수에 1씩 빼기

```python
result += (n-1)
print(result)
```
