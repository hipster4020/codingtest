# *큰 수의 법칙*

# 

##### 값 입력

```python
n, m, k = map(int, input().split())
```

```python
data = list(map(int, input().split()))
```

```python
print(f"n : {n}")
print(f"m : {m}")
print(f"k : {k}")
print(f"data : {data}")
```

##### 정렬 및 1, 2 순위 추출

```python
data.sort()
data
```

```python
first = data[len(data) - 1]
second = data[len(data) - 2]
```

```python
print(f"first : {first}")
print(f"second : {second}")
```

##### 횟수 계산

```python
count = int(m / (k+1)) * k
print(f"count : {count}")
```

```python
count += m % (k+1)
print(f"count : {count}")
```

##### 더하기

```python
result = 0
result += (count) * first
result += (m - count) * second
print(f"result : {result}")
```
