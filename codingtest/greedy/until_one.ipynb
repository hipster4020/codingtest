```python
# n = 25 #  
# k = 5 #  
n = 23
k = 3
```

```python
def main(n, k):
    count = 0
    while n >= k:
        if n % k == 0:
            n //= k
            count += 1
        else:
            n -= 1
            count += 1
    return count
```

```python
result = main(n, k)
```
