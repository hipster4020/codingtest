# 1이 될 때까지


```python
# n = 25 # 나누어야하는 수
# k = 5 # 나눌 수
n = 23
k = 3
```

```python
def main(n, k):
    count = 0
    while n >= k:
        if n % k == 0:
            n //= k
            count += 1
        else:
            n -= 1
            count += 1
    return count
```

```python
result = main(n, k)
```

```python
print(result)
```
