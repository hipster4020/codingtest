# 거스름돈


```python
n = 1260
```

```python
def main(n):
    count = 0
    coins = [500, 100, 50, 10]
    
    for coin in coins:
        count += n // coin
        print(f"coin : {coin}")
        n %= coin
        print(f"n : {n}")
    return count
```

```python
result = main(n)
```

```python
print(result)
```
