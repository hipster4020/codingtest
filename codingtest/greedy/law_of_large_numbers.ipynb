# 큰 수의 법칙



```python
n = 5    # 배열 크기
m = 8    # 숫자가 더해지는 횟수
k = 3    # 특정 번호가 k번을 연속해서 k번 더할 수 없음
data = [2, 4, 5, 4, 6]
```

```python
def main(n, m, k, data):
    data.sort(reverse=True)
    first = data[0]
    second = data[1]

    result = 0

    while True:
        for _ in range(k):
            if m == 0:
                break
            result += first
            m -= 1
        if m == 0:
            break
        result += second
        m -= 1

    return result
```

```python
result = main(n, m, k, data)
```
