```python
n = 5    #  
m = 8    #   
k = 3    #   k  k   
data = [2, 4, 5, 4, 6]
```

```python
def main(n, m, k, data):
    data.sort(reverse=True)
    first = data[0]
    second = data[1]

    result = 0

    while True:
        for _ in range(k):
            if m == 0:
                break
            result += first
            m -= 1
        if m == 0:
            break
        result += second
        m -= 1

    return result
```

```python
result = main(n, m, k, data)
```
